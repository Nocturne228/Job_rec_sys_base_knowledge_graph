```python
import pandas as pd
import numpy as np
import re
import jieba
```

```python
df = pd.read_pickle('processed_data.pkl')
```

```python
# 加载停用词表
stop_words = open('../../extractor/hit_stopwords.txt', 'r', encoding='utf-8').readlines() 
stop_words = [w.strip() for w in stop_words]

def preprocess(text):
    if isinstance(text, str):
        # 分词
        words = jieba.lcut(text)
        # words = jieba.cut(text, cut_all=True)
        
        # 去停用词
        words = [w for w in words if w not in stop_words]
        
        # 去标点符号
        words = [re.sub(r'[^\w\s]', '', w) for w in words]
        
        # 去空格
        words = [w.strip() for w in words]
        
        # 去空行和单字
        words = [w for w in words if w and len(w) > 1]
        
        return words
    else:
        return []
    
```

```python
data = df
# 增加'关键词'列,并逐行加入分词结果  
data['招聘关键词'] = ''
data['要求关键词'] = ''


for index, row in data.iterrows():
    text = row['招聘描述']
    words = preprocess(text)
    data.at[index, '招聘关键词'] = ",".join(words)
    
    
for index, row in data.iterrows():
    text = row['要求']
    words = preprocess(text)
    data.at[index, '要求关键词'] = ",".join(words)


data
```

```python
city_list = pd.read_csv('china_city_list.csv', encoding='gbk')
cities = city_list['City'].to_list()
cities
```

```python
for index, row in data.iterrows():
    for city in cities:
        if city in row['地区']:
            data.at[index, '地区'] = city
            break
            
data
```

```python
data.to_pickle('process1.pkl')
```
